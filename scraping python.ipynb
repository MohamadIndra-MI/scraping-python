```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"

tables = pd.read_html(url)
```

```python
print(f"Ada {len(tables)} tabel yang ditemukan.")
```

```python
df = tables[0]
```

```python
print(df.head())
```

```python
df.to_csv("hasil_wikipedia.csv", index=False)
```

```python
df = df.reset_index(drop=True)
```

```python
# Gabungkan multi-level kolom menjadi satu baris header
df.columns = [' '.join(col).strip() if isinstance(col, tuple) else col for col in df.columns]

# Lalu simpan ke Excel
df.to_excel("hasil_wikipedia.xlsx", index=False)
```
